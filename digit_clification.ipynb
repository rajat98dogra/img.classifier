```python
x_train, x_test = x[:600],x[600:]
y_train, y_test = y[:600],y[600:]
```

```python
import skimage.io as read
import skimage.io as save
import skimage.io as show
import numpy as np
shuffle_index = np.random.permutation(600)
x_train, y_train = x_train[shuffle_index], y_train[shuffle_index]
```

```python
from sklearn import svm
```

```python
clf = svm.SVC(kernel="linear")
```

```python
from sklearn.metrics import accuracy_score
print("model accuracy:",accuracy_score(y_test,prdt))
```
